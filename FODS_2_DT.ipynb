df.head() - Displays the first 5 rows of the DataFrame.

df.tail() - Displays the last 5 rows.

df.info() -  Provides a concise summary of the DataFrame.

df.shape -  Returns the dimensions of the DataFrame.

df.describe() -  Provides descriptive statistics.

df['column_name'].value_counts() -  Returns counts of unique values in a column.

df.isnull().sum() -  Returns the number of missing values per column.

df.dropna() -  Removes rows with missing values.

df.groupby('column_name').sum() -  Groups data by a column and computes the sum for each group.
